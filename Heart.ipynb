# Data Preprocessing

## Splitting Feature and Target

```python
X = data.drop(columns=['death_event'])
y = data['death_event']
```

## Handling Imbalance Dataset

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
rus = RandomOverSampler(random_state = 0)
X, y = rus.fit_resample(X, y)
```

## Splitting Training and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Feature Scaling

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train.iloc[:, [0, 2, 4, 6, 7, 8, 11]] = sc.fit_transform(X_train.iloc[:, [0, 2, 4, 6, 7, 8, 11]])
X_test.iloc[:, [0, 2, 4, 6, 7, 8, 11]] = sc.fit_transform(X_test.iloc[:, [0, 2, 4, 6, 7, 8, 11]])
```

```python
X_train
```

# Modeling

## Template

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score

def modeling(model, name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = cross_val_score(model, X_train, y_train, cv=10)
    print(f'{name} Evaluation')
    print('Model Accuracy: ', accuracy_score(y_test, y_pred))
    print('Model Precission: ', precision_score(y_test, y_pred))
    print('Model Recall: ', recall_score(y_test, y_pred))
    print('Cross Validation (10): ', score.mean())
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
linear = LogisticRegression()
modeling(linear, 'Logistic Regression')
```

## SVM

### Linear Kernel

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear')
modeling(svm, 'SVM')
```

### Rbf kernel

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
modeling(svm, 'SVM')
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=60)
modeling(forest, 'Random Forest')
```

## XGBoost

```python
from xgboost import XGBClassifier
xgboost = XGBClassifier()
modeling(xgboost, 'XGBoost')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
modeling(knn, 'KNN')
```
