## Relationship between Features

### Loan Sanction Amount  (USD) Distribution -> Target Feature

<img src = 'dist.png' style = 'width:1000px'>

```python
fig, ax = plt.subplots(1, 2, figsize = (18, 8))
sns.histplot(data = data, x ='Loan Sanction Amount (USD)', bins = 30, kde = True, ax = ax[0])
ax[0].set_title('Histogram Distribution of Loan Sanction Amount (USD)')
ax[1].set_title('BoxPlot Distribution of Loan Sanction Amount (USD)')
sns.boxplot(data = data, x = 'Loan Sanction Amount (USD)')
plt.show()
```

Based on the graph, we can see <b>Loan Sanction Amount(USD) </b>is <b> Positively (rigt) skewed distribution </b> and has many outliers from the boxplot.<br>
Because some data have their Loan Sanction Amount (USD) missing, we need to drop it because we could not make an assumption of the target result

### ID & Name

```python
data['Customer ID'].value_counts()
```

```python
data['Name'].value_counts()
```

CustomerID and Name only <b>represent the rows</b > within the dataset, thus we do not need it in our Machine Learning Prediction.

```python
data.drop(columns=['Customer ID', 'Name'], inplace = True)
```

### Gender

```python
fig, ax = plt.subplots(1,2 , figsize = (18, 8))
data['Gender'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%', ax = ax[0])
sns.pointplot(data = data, x ='Gender', y = 'Loan Sanction Amount (USD)',ax = ax[1])
ax[1].set_title('Loan Sanction Amount by Gender')
ax[0].set_title('Distribution of Gender')
plt.show()
```

From the graph, we can see that while many Customers are male, the Loan Sanction Amount is slightly bigger from Female

### Age

```python
fig, ax = plt.subplots(1, 2, figsize =(18, 8))
sns.histplot(data = data, x = 'Age', ax = ax[0])
sns.boxplot(data = data, x ='Age')
plt.show()
```

Our customer has mean age of around 40- with the youngest one around 25 and the oldest are around 55, we also got Binomial Distribution from the age

### Income

```python
fig, ax = plt.subplots(1, 2, figsize = (18, 8))
sns.scatterplot(x='Income (USD)', y='Loan Sanction Amount (USD)', data=data)
sns.boxplot(data = data, x = 'Income (USD)', ax = ax[0])
```

```python
income = np.log(data['Income (USD)'])
loan = np.log(data['Loan Amount Request (USD)'])

plt.scatter(x=income, y=loan)
plt.xlabel('Log of Income (USD)')
plt.ylabel('Log of Loan Amount Requested (USD)')
plt.title('Logarithmic Scaling of Income vs Loan Amount')
plt.show()
```
