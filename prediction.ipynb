# Feature Selection

## Splitting Features and Target

```python
X = data.drop(columns= ['Loan Sanction Amount (USD)'])
y = data['Loan Sanction Amount (USD)']
```

## Train and Test Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train.iloc[:, [0, 1, 2, 3, 4, 6, 9]] = sc.fit_transform(X_train.iloc[:, [0, 1, 2, 3, 4, 6, 9]])
X_test.iloc[:, [0, 1, 2, 3, 4, 6, 9]] = sc.transform(X_test.iloc[:, [0, 1, 2, 3, 4, 6, 9]])
```

# Modeling

```python
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import cross_val_score
def modeling(model):
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    print('Model R2Score: ', r2_score(prediction, y_test))
    score = cross_val_score(model, X_train, y_train, cv=10, scoring='accuracy')
    print('Cross Validation Score: ', score.mean())
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
modeling(linear)
```
