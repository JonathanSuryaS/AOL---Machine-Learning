# Exploratory Data Analysis

## Relationship between Features

### satisfaction -> Target feature

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))
data['satisfaction'].value_counts().plot(kind = 'pie', ax = ax[0], autopct = '%1.1f%%', explode = [0, 0.1])
ax[0].set_ylabel('')
ax[0].set_title('Satisfaction Distribution (Pie chart)')
ax[0].set_title('Satisfaction Distribution (Count plot)')
sns.countplot(data = data, x = 'satisfaction', hue = 'satisfaction')
plt.show()
```

However, as we need to analyze the relation between features and target, we need to immediately encode our target feature first, as satisfaction == 1, and 0 for others.

```python
data.loc[data['satisfaction']=='satisfied', 'satisfaction'] = 1
data.loc[data['satisfaction']=='neutral or dissatisfied', 'satisfaction'] = 0
```

### Unnamed: 0

```python
data.drop(columns = ['Unnamed: 0'], inplace = True)
```

### Gender

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))
data['Gender'].value_counts().plot(kind = 'pie', ax = ax[0], autopct = '%1.1f%%', explode = [0, 0.1])
ax[0].set_ylabel('')
ax[0].set_title('Gender Distribution (Pie chart)')
ax[1].set_title('Pointplot Gender vs Satisfaction')
sns.pointplot(data = data, x = 'Gender', y = 'satisfaction')
plt.show()
```

Based from our graph, Male customer is easier to satisfy as it have higher chance for satisfied rather than Female 

### Customer Type
Type of Customer -> Loyal / Disloyal

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))
data['Customer Type'].value_counts().plot(kind = 'pie', ax = ax[0], autopct = '%1.1f%%', explode = [0, 0.1])
ax[0].set_ylabel('')
ax[0].set_title('Customer Type Distribution (Pie chart)')
ax[1].set_title('Pointplot Customer Type vs Satisfaction')
sns.pointplot(data = data, x = 'Customer Type', y = 'satisfaction')
plt.show()
```

From the graph above, we can see that it's easier to satisfy loyal customer than disloyal customer, as loyal customer usually means that this customer is used to buy airline ticket from a single company many times (loyal).

### Age
Age of Customer

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 8))
sns.histplot(data = data, x = 'Age', ax = ax[0], bins = 30, kde = True)
ax[0].set_ylabel('')
ax[0].set_title('Customer Age Distribution (Histogram)')
ax[1].set_title('Customer Age Distribution (Boxplot)')
sns.boxplot(data = data, x = 'Age')
plt.show()
```

The distribution of an age is mostly unimodal, where the data does not have an outlier as seen in the boxplot, there is no white dot or an outliers. it said that, the mean age of customer who use airline service is 40 y.0

```python
plt.figure(figsize = (18, 8))
sns.histplot(data = data, x = 'Age', bins = 30, hue = 'satisfaction', kde = True)
plt.title('Customer Age vs Satisfaction')
plt.show()
```

We can see from the graph that customer with age around 40 to 60 y.o are easier to satisfy as the frequency gradually increase for this specific range of age.
