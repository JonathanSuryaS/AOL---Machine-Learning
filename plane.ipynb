# Data Preprocessing

## Handling Missing Value

Because the missing value is even less than 10%, we can simply drop the NULL Entries

```python
new_data.dropna(inplace= True)
(new_data.isnull().sum()).sum()
```

## Splitting Features and Target 

```python
X = new_data.drop(columns=['satisfaction'])
y = new_data['satisfaction']
```

```python
X
```

## Encoding Categorical Variable

```python
X = pd.get_dummies(X, drop_first=True)
X
```

```python
X.dtypes
```

## Splitting Train and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train.iloc[:, [0, 1, 14]] = sc.fit_transform(X_train.iloc[:, [0, 1, 14]])
X_test.iloc[:, [0, 1, 14]] = sc.transform(X_test.iloc[:, [0, 1, 14]])
```

```python
X_train
```

# Modeling

## Template

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score

def modeling(model, name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = cross_val_score(model, X_train, y_train, cv=10)
    print(f'{name} Evaluation')
    print('Model Accuracy: ', accuracy_score(y_test, y_pred))
    print('Model Precission: ', precision_score(y_test, y_pred))
    print('Model Recall: ', recall_score(y_test, y_pred))
    print('Cross Validation (10): ', score.mean())
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
```

```python
from sklearn.linear_model import LogisticRegression
linear = LogisticRegression()
modeling(linear, 'Logistic Regression')
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=60)
modeling(forest, 'Random Forest')
```

```python
y_pred = forest.predict(X_test)
comparison = np.column_stack((y_test, y_pred))

# Print the comparison
print("Actual   Predicted")
for actual, predicted in comparison:
    print(actual, "    ", predicted)
```

```python
from xgboost import XGBClassifier
xgboost = XGBClassifier()
modeling(xgboost, 'XGBoost')
```

# Saving the best model

Since we can conclude, that the best model is represented by Random Forest Classification, we will save and use RandomForest Classification model for further use.

## Feature Importance according RandomForestClassifier Model

```python
importance = forest.feature_importances_

for i,v in enumerate(importance):
    print('Feature: %d, Score: %.5f' % (i,v))
# plot feature importance
plt.figure(figsize=(15,15))
plt.bar(X.columns, importance)
plt.xticks(rotation=90)

plt.show()
```
